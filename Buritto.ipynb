```python
import pandas as pd
import numpy as np
import atoti as tt
import re
```

```python
df = pd.read_csv('burritos_01022018.csv') 
```

```python
df.columns
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns = [re.sub("([\(\[]).*?([\)\]])","",x).strip() for x in df.columns]
df.columns = [x.replace(':','_').strip() for x in df.columns]
```

```python
df.columns
```

```python
df.isnull().any()
df.isnull().sum()/df.shape[0]
```

```python
session = tt.Session(
    port=9000,
    user_content_storage="./content",
)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
buritto_vars_norm = df.loc[:,['Circum','Volume','Length','Mass','Cost']]
bnorms = scaler.fit_transform(buritto_vars_norm)*10

df[['Circum_norm','Volume_norm','Length_norm','Mass_norm','Cost_norm']] = bnorms
```

```python
df.columns
```

```python
burito_variables = pd.melt(df.reset_index(),id_vars = ['Location','Burrito'],value_vars = ['Circum_norm','Volume_norm','Length_norm','Mass_norm','Cost_norm'],)
burito_variables
```

```python
buritto_table = session.read_pandas(df,table_name = "burritos")
```

```python
buritto_table.head()
```

```python
cube = session.create_cube(buritto_table)
```

```python
h = cube.hierarchies
l = cube.levels
m = cube.measures
```

```python
m["five"] = 5
m["lenXwrap"] = m["Length.MEAN"] * m["Wrap.MEAN"]
```

```python
session.visualize("exploration 1")
```

```python
session.visualize("scatter plot neighbourhood yelp google")
```

```python
buritto_var_table = session.read_pandas(burito_variables,table_name = "buritto_variables_x",keys = ["Location","Burrito","variable","value"],)
```

```python
buritto_table.join(buritto_var_table)
```

```python
m["value"]=tt.agg.mean(buritto_var_table["value"])
m["aggvalue"]=tt.agg.mean(
    m["value"], scope = tt.OriginScope(l["Location"],l["Burrito"],l["variable"])
)
```

```python
session.visualize("Radar chart final")
```
